# matplotlib pyplot
- pyplot is sub module of mstplotlib and acess through plt

```python

import pandas as pd
import matplotlib.pyplot as plt 
import numpy as np
x=np.array([0,6,8])
y =np.array([50,250,800])
plt.plot(x,y)
plt.title("My First Graph of X and Y")
plt.show()
```

**IF I WANT TO SHOW GRAPH WIYHOUT LINE**
- IN PYTHON WHEN WEE PLOTT A  VALUE LIKE X AND Y IT SHOW THOSE VALUE IN LINE 

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x=np.array([0,10])
y=np.array([10,100])
plt.plot(x,y,'o')   
plt.title("GRAPH WITHOUT LINE")
plt.show()

```

# MARKER

**you want to amrk each point with diffferent shape**
- we use marker

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x =np.array([1,3,5,6,7])
y=np.array([2,4,6,8,10])
plt.plot(x,y,marker='o')
plt.title("GRAPH WITH MARKER")
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
x =np.array([1,3,5,6,7])
y=np.array([2,4,6,8,10])
plt.plot(x,y,marker='*')
plt.title("GRAPH WITH MARKER")
plt.show()
```
